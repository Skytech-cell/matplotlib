```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
Avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Tempretutre')
plt.title('Weather')

plt.plot(day,max_t)
plt.plot(day,min_t)
plt.plot(day,Avg_t)
plt.legend(['Max','Min','Avg'],loc='best',shadow=True,fontsize='small')
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.ylabel('Revenue')
plt.title('Tech Stocks')
plt.bar(ypos,revenue,label='Revenue')
plt.bar(ypos,profit,label='Profit')
plt.legend()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel('Revenue')
plt.title('Tech Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
plt.yticks(ypos,company)
plt.title('Tech Stocks')
plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2,profit,label='Profit')
plt.legend()
```

```python
import matplotlib.pyplot as plt
blood_sugar=[113,85,90,150,149,88,93,115,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150], color='g',rwidth=0.95)
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,80,77,82,129]
blood_sugar_women=[67,98,59,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('Sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150], color=['green','orange'],rwidth=0.95,label=['men','women'],orientation='horizontal')
plt.legend()
```

```python
import matplotlib.pyplot as plt
exp_vals=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','Phone/Internet Bill','car','other Utilities']
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%',shadow=True, explode=[0,0.1,0,0,0.1],startangle=45)
plt.show()
```
